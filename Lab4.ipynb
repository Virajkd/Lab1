```python
#Question 1
import pandas as pd
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
filepath='/content/VSP - Java.xlsx'
df=pd.read_excel(filepath)

X = df.head(150)[['Header_and_Main_declaration','Incomprehensible_Code','Comprehensible_code_with_logical_errors','Comprehensible_code_with_syntax_errors','Correct_Code_and_Output'	]].values
y = np.where(df.head(150)['Final_Marks'] <= 5, 0, 1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(X_train, y_train)


print("X_train:\n", X_train)
print("X_test:\n", X_test)
print("y_train:\n", y_train)
print("y_test:\n", y_test)
y_train_pred = neigh.predict(X_train)

y_test_pred = neigh.predict(X_test)

print("Training Data Evaluation")
train_conf_matrix = confusion_matrix(y_train, y_train_pred)
print("Confusion Matrix:\n", train_conf_matrix)
train_class_report = classification_report(y_train, y_train_pred, target_names=["Class 0", "Class 1"])
print("Classification Report:\n", train_class_report)

print("Test Data Evaluation")
test_conf_matrix = confusion_matrix(y_test, y_test_pred)
print("Confusion Matrix:\n", test_conf_matrix)
test_class_report = classification_report(y_test, y_test_pred, target_names=["Class 0", "Class 1"])
print("Classification Report:\n", test_class_report)

train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)

print(f"Training Accuracy: {train_accuracy}")
print(f"Test Accuracy: {test_accuracy}")

if train_accuracy > test_accuracy and test_accuracy < 0.7:
    print("The model might be overfitting.")
elif train_accuracy < test_accuracy and test_accuracy < 0.7:
    print("The model might be underfitting.")
else:
    print("The model seems to be fitting well (regular fit).")
```

```python
#Question 2

from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error, r2_score

path_of_file = '/content/Lab Session Data.xlsx'
price = pd.read_excel(path_of_file, sheet_name='IRCTC Stock Price')
price['Predicted_Price'] = price['Price'].shift(1)
price = price.dropna()

realprice = price['Price'].values
assumedprice = price['Predicted_Price'].values

mse = mean_squared_error(realprice, assumedprice)
rmse = np.sqrt(mse)
mape = mean_absolute_percentage_error(realprice, assumedprice)
r2 = r2_score(realprice, assumedprice)

print("mse is:",mse)
print("rmse is:",rmse)
print("mape is:",mape)
print("r2 score is:",r2)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV

# Load the first 100 rows from the Excel file
excel_data = pd.read_excel('/content/VSP - Java.xlsx', nrows=100)

# Select the first two columns as features X and Y
X = excel_data.iloc[:, 0].values
Y = excel_data.iloc[:, 1].values
final_marks = excel_data['Final_Marks'].values

# Check the distribution of Final_Marks
print("Final_Marks Distribution:")
print(pd.Series(final_marks).describe())
plt.hist(final_marks, bins=20, edgecolor='black')
plt.title('Distribution of Final_Marks')
plt.xlabel('Final_Marks')
plt.ylabel('Frequency')
plt.show()

# Adjust the threshold if needed
threshold = 5  # Adjust this based on the distribution if necessary

# Assign classes based on the adjusted threshold
data = pd.DataFrame({
    'X': X,
    'Y': Y,
    'Final_Marks': final_marks
})
data['Class'] = np.where(data['Final_Marks'] > threshold, 1, 0)  # Class 1 if marks > threshold, else class 0

# Check the class distribution
print("Class Distribution:")
print(data['Class'].value_counts())

# Sample 20 data points for training
train_data = data.sample(n=20, random_state=42)
test_data = data.drop(train_data.index)

# Plot the training data
plt.figure(figsize=(8, 6))
for class_value, color in zip([0, 1], ['blue', 'red']):
    subset = train_data[train_data['Class'] == class_value]
    plt.scatter(subset['X'], subset['Y'], c=color, label=f'Class {class_value}', edgecolors='k')

plt.title('Training Data Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()

# Create a grid of test points
x_min, x_max = data['X'].min() - 1, data['X'].max() + 1
y_min, y_max = data['Y'].min() - 1, data['Y'].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1), np.arange(y_min, y_max, 0.1))
grid_points = np.c_[xx.ravel(), yy.ravel()]

# Train kNN with k=3 and classify the grid points
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(train_data[['X', 'Y']], train_data['Class'])
predicted_classes = knn.predict(grid_points)
predicted_classes = predicted_classes.reshape(xx.shape)

# Plot the classification results
plt.figure(figsize=(10, 8))
plt.contourf(xx, yy, predicted_classes, alpha=0.5, cmap='coolwarm')
for class_value, color in zip([0, 1], ['blue', 'red']):
    subset = train_data[train_data['Class'] == class_value]
    plt.scatter(subset['X'], subset['Y'], c=color, label=f'Class {class_value}', edgecolors='k')

plt.title('kNN Classification with k=3')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()

# Perform GridSearchCV to find the best k value
param_grid = {'n_neighbors': np.arange(1, 21)}
grid_search = GridSearchCV(KNeighborsClassifier(), param_grid, cv=5)
grid_search.fit(train_data[['X', 'Y']], train_data['Class'])

best_k = grid_search.best_params_['n_neighbors']
print(f'The best value for k is: {best_k}')
```
